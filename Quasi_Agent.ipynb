```python
import os
from google.colab import userdata

api_key = userdata.get('OPENAI_API_KEY')
os.environ['OPENAI_API_KEY'] = api_key

from litellm import completion
from typing import List, Dict
def model_discussion(messages: List[Dict]) -> str:
  response = completion(
      messages = messages,
      model = "openai/gpt-4o",
      max_tokens = 1024
  )
  return response.choices[0].message.content

user_input1 = input("what would you like to create today?")
user_input2 = input("is there anything else you would like... say a documentation")
user_input3 = input("any final requests you would like... say a unittest")

messages = [
    {"role": "system", "content": "you are to act as a software engineer and meant to provide a  python version of a function if asked to by a user"},
    {"role": "user", "content": user_input1}, # Use user_input1
]

response1 = model_discussion(messages) # Call the function

print(response1)

messages2 = [
    {"role": "system", "content": "you are to act as a software engineer and meant to provide a  python version of a function if asked to by a user"},
    {"role": "user", "content": user_input2},
    {"role": "assistant", "content": response1} # Use response1
]

response2 = model_discussion(messages2) # Call the function

print(response2)

messages3 = [
    {"role": "system", "content": "you are to act as a software engineer and meant to provide a  python version of a function if asked to by a user"},
    {"role": "user", "content": user_input3},
    {"role": "assistant", "content": response2} # Use response2
]

response3 = model_discussion(messages3) # Call the function

print(response3)
```
